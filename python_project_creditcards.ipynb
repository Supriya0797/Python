# The relationship Between the Variables and Default
We'll research how features and credit card default are related. Even if it's not conclusive, this will help me grasp the data. This will also aid me in deciding how to handle features whose values deviate from those specified in the data description.

```python
from matplotlib.pyplot import figure
import matplotlib as plt
```

```python
def countplot_credit(columns):
    figure(figsize=(15,10))
    plot=sns.countplot(x=columns,data=credit_dataframe)
    total=len(credit_dataframe['default payment next month'])
    
    for p in plot.patches:
        
        plot.annotate('{:.1f}%'.format(100*p.get_height()/total), (p.get_x(), p.get_height()))
    plot.yaxis.set_ticks(np.linspace(0, total, 11))
    plot.set_yticklabels(map('{:.1f}%'.format, 100*plot.yaxis.get_majorticklocs()/total))
    plt.pyplot.show()
```

```python
def countplot_default(columns,classes):
    figure(figsize=(15,10))
    total=len(credit_dataframe[classes])
    plot=sns.countplot(x=columns,data=credit_dataframe,hue=classes)
    for p in plot.patches:
        plot.annotate('{:.1f}%'.format(100*p.get_height()/total), (p.get_x(), p.get_height()))
    plot.yaxis.set_ticks(np.linspace(0, total, 11))
    plot.set_yticklabels(map('{:.1f}%'.format, 100*plot.yaxis.get_majorticklocs()/total))
    plt.pyplot.show()
```

```python
def limit_balance_group(r):
    if r['LIMIT_BAL'] > 10000 and r['LIMIT_BAL'] <= 80000:
        return 'level1'
    elif r['LIMIT_BAL'] > 90000 and r['LIMIT_BAL'] <=170000:
        return 'level2'
    elif r['LIMIT_BAL'] > 180000 and r['LIMIT_BAL'] <=260000:
        return 'level3'
    elif r['LIMIT_BAL'] > 270000 and r['LIMIT_BAL'] <=340000:
        return 'level4'
    elif r['LIMIT_BAL'] > 350000 and r['LIMIT_BAL'] <=430000:
        return 'level5'
    elif r['LIMIT_BAL'] > 440000 and r['LIMIT_BAL'] <=520000:
        return 'level6'
    elif r['LIMIT_BAL'] > 530000 and r['LIMIT_BAL'] <=610000:
        return 'level7'
    elif r['LIMIT_BAL'] > 620000 and r['LIMIT_BAL'] <=700000:
        return 'level8'
    else:
        return 'level9'
credit_dataframe['limit_balance_group'] = credit_dataframe.apply(lambda x: limit_balance_group(x),axis = 1)
```

# Encoding Categorical Data

Three categorical demographic variables were present in the original dataset (sex, marriage and education). Given that changes in value might result in more or less education, among them, education has an ordinal structure. Sex only has two classes, making it one-hot encoded. I will now encode the marital variable in one hot.

```python
i = credit_dataframe.columns.get_loc("MARRIAGE")

left = credit_dataframe.iloc[:, :i]
marriage = pd.get_dummies(credit_dataframe["MARRIAGE"], prefix="MARRIAGE", drop_first=True)
right = credit_dataframe.iloc[:, i + 1 :]

credit_dataframe = pd.concat([left, marriage, right], axis=1)
```

```python
X = credit_dataframe[['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE_1', 'MARRIAGE_2',
       'MARRIAGE_3', 'AGE', 'PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5',
       'PAY_6', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4',
       'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3',
       'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']]

X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=1)
```

```python
KNN.fit(X_train,y_train)
```

```python
pred_KNN = KNN.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred_KNN))
```

```python
print(classification_report(y_test,pred_KNN))
```

```python
errrate = []

for i in range(1,40):
    
    KNN = KNeighborsClassifier(n_neighbors=i)
    KNN.fit(X_train,y_train)
    pred_KNNi = KNN.predict(X_test)
    errrate.append(np.mean(pred_KNNi != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),errrate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

Here we can see that that after arouns K>18 the error rate just tends to hover around <0.21 Let's retrain the model with that and check the classification report!

```python
KNN = KNeighborsClassifier(n_neighbors=1)

KNN.fit(X_train,y_train)
pred = KNN.predict(X_test)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
KNN = KNeighborsClassifier(n_neighbors=15)

KNN.fit(X_train,y_train)
pred = KNN.predict(X_test)

print('WITH K=18')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
